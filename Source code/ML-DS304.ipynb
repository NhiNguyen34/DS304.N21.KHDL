```python
# Load dataset
raw_data = torch.load('/content/drive/MyDrive/UIT/Hien/Data/aptos2019-raw.pt')

X_train, X_test_, y_train, y_test_ = train_test_split(raw_data.tensors[0], raw_data.tensors[1], test_size=0.33, random_state=42)
X_test, X_valid, y_test, y_valid = train_test_split(X_test_, y_test_, test_size=0.33, random_state=42)

# Scale data
X_train = (X_train/255).permute(0, 2, 3, 1).numpy()
X_valid = (X_valid/255).permute(0, 2, 3, 1).numpy()
X_test = (X_test/255).permute(0, 2, 3, 1).numpy()

y_train = y_train.numpy()
y_valid = y_valid.numpy()
y_test = y_test.numpy()
```

```python
# Turn to grayscale
X_train = 0.2989*X_train[:,:,:,0] + 0.5870*X_train[:,:,:,1] + 0.1140*X_train[:,:,:,2]
X_test = 0.2989*X_test[:,:,:,0] + 0.5870*X_test[:,:,:,1] + 0.1140*X_test[:,:,:,2]

# Flatten though image
X_train = X_train.reshape(2453, 50176)
X_test = X_test.reshape(810, 50176)
```

```python
from sklearn.svm import SVC
from timeit import default_timer as timer

# Fit Model
start = timer()
model = SVC(gamma='auto').fit(X_train, y_train)
stop = timer()
print(f'Total fit time: {stop - start}')

# Predict Model
start = timer()
y_pred = model.predict(X_test)
stop = timer()
print(f'Total predict time: {stop - start}')
```

```python
print(f1_score(y_test, y_pred, average='micro'))
```
